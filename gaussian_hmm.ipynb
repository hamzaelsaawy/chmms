### project trained transition matrix to each trajectory

```python
As = np.zeros((n_trajs, n_components, n_components))
for s, (i, j) in enumerate(hmm.iter_from_X_lengths(X, traj_lens)):
    # s is the trajectory number
    framelogprob = model._compute_log_likelihood(X[i:j])
    _, fwdlattice = model._do_forward_pass(framelogprob)
    bwdlattice = model._do_backward_pass(framelogprob)
    # skip the last timestep thingy
    temp = np.zeros((n_components, n_components))
    for t in range(j-i-1):
        temp = fwdlattice[t, :, np.newaxis] + model.transmat_ + bwdlattice[t+1] + framelogprob[t+1]
        temp -= logsumexp(temp)
        As[s] += np.exp(temp)
    As[s] /= As[s].sum(0, keepdims=True)
```

### distance between all trajectories

```python
D = np.empty((n_trajs, n_trajs))

# number of trajectories around a trajectory to consider 
look_ahead = 150

for i in range(n_trajs):
    for j in range(i, n_trajs):
        D[i, j] = np.log(n_components) * np.sum(np.sqrt(As[i] * As[j]))
        D[j, i] = D[i, j]
```

### find pairs

```python
# max lane difference
lane_max = 1 
# max distance between cars
d_max = 10 
# maximum number of trajectories away
max_interaction = 300
# max length of interaction (time steps)
min_length = 20
# indices of cars that are interacting
traj_pairs = []

traj_idx = list(hmm.iter_from_X_lengths(X, traj_lens))
```

```python
start_overlap = 0
end_overlap = 0
for i in range(n_trajs):
    i_idx = traj_idx[i]
    traj_i = X_full.loc[range(*i_idx)]
    
    if i % 75 == 0:
        print(i)
    
    max_j = min(i + max_interaction, n_trajs-1)
    for j in range(i + 1, max_j):
        j_idx = traj_idx[j]
        traj_j = X_full.loc[range(*j_idx)]

        # first frame (time step) that both trajectories are defined
        start_frame = max(traj_i.frame.iloc[0], traj_j.frame.iloc[0])
        # earliest frame (time step) that either trajectory ends
        # plus 1 to *include* for stupid python indexing
        end_frame = 1 + min(traj_i.frame.iloc[-1], traj_j.frame.iloc[-1])
        
        rng = range(start_frame, end_frame)
        if len(rng) < min_length:
            continue

        # offset into each trajectory where they overlap
        i_offset = start_frame - traj_i.frame.iloc[0]
        j_offset = start_frame - traj_j.frame.iloc[0]

        start_overlap = 0
        for k in range(len(rng)):
            si = traj_i.iloc[k + i_offset]
            sj = traj_j.iloc[k + j_offset]

            d = sj[['x', 'y']] - si[['x', 'y']]
            ds = np.sqrt(np.sum(d ** 2))
            dlane = np.abs(si.lane - sj.lane)

            # cars interacting?
            if (ds <= d_max) and (dlane <= lane_max):
                if start_overlap == 0:
                    start_overlap = k
                end_overlap = k+1

        len_overlap = end_overlap - start_overlap + 1
        if (start_overlap != 0) and ( len_overlap >= min_length ):
            i_info = (i, (i_offset + start_overlap, i_offset + end_overlap))
            j_info = (j, (j_offset + start_overlap, j_offset + end_overlap)) 
            traj_pairs.append( (i_info, j_info) )
```
